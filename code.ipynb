```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('./Machine-Learning-Final-Project/CS559_S23_Competition_Data/train_.csv')
dataset
```

```python
#Preprocessing the Data
#Removing Info from the dataset that are not important
dataset= dataset.drop(['PassengerId', 'Name','Unnamed: 0'], axis=1)
#drop passengerId and Name and Unnamed:0 because we know that the passengers that are 
#transported are definetely not chosen by their unique identification or name
```

```python
#check for missing values in the dataset
dataset.isnull().sum().sort_values(ascending=False)
```

```python
#Since there are missing values, fill them in with a value that does not affect
#the graph like 0
dataset[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] = dataset[['VIP', 'CryoSleep', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(value=0)
```

```python
#change the data values that are booleans to something else to represent them
#like 1 and 0s. 
dataset["Transported"] = dataset["Transported"].astype(int)
dataset['VIP'] = dataset['VIP'].replace({True: 1, False: 0}).astype(int)
dataset['CryoSleep'] = dataset['CryoSleep'].replace({True: 1, False: 0}).astype(int)
```

```python
#Cabin has three numbers Deck/Cabin Num/Side so we would 
#split the dataset into three new columns to make it easier
#to work with
dataset[["Deck", "Cabin_num", "Side"]] = dataset["Cabin"].str.split("/", expand=True)
#Remove Cabin as we don't need it anymore
dataset = dataset.drop('Cabin', axis=1)
```

```python
#Filling in Age with the mean of the ages as it is a continuous variable
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
```

```python
#Filling in HomePlanet with the mode of the homeplanets as it is a catergorial variable
dataset['HomePlanet'].value_counts()
#Earth has the most so we'll fill in the blanks with that
dataset['HomePlanet'].fillna('Earth',inplace=True)
```

```python
#Filling in Destination with the mode of the destinations as it is a catergoiral variable
dataset['Destination'].value_counts()
#TRAPPIST-1e has the most so we'll fill in the blanks with that
dataset['Destination'].fillna('TRAPPIST-1e',inplace=True)
```

```python
#Filling in Deck with the mode of the deck as it is a catergoiral variable
dataset['Deck'].value_counts()
#F has the most so we'll fill in the blanks with that 
dataset['Deck'].fillna('F',inplace=True)
```

```python
#Filling in Side with the mode of the side as it is a catergoiral variable
dataset['Side'].value_counts()
#S has the most so we'll fill in the blanks with S
dataset['Side'].fillna('S',inplace=True)
```

```python
#Since there are missing values, fill them in with a value that does not affect
#the graph like 0
dataset[['Cabin_num','RoomService']] = dataset[['Cabin_num','RoomService']].fillna(value=0)
dataset["Cabin_num"] = dataset["Cabin_num"].astype(int)
dataset["RoomService"] = dataset["RoomService"].astype(int)
```

```python
#check dataset again to see if there are no more null values
dataset.isnull().sum().sort_values(ascending=False)
```
